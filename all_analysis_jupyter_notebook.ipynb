# PM1 OPC N2 1 VS GRIMM

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
y = df['PM1_OPC_N2_1'].values.reshape(-1,1)
X = df['PM1_GRIMM'].values.reshape(-1,1)
regr.fit(X, y)

a=(regr.coef_[0])
b=(regr.intercept_)
print(a,b)
```

```python
#RMSE Calculation

from sklearn import linear_model
regr = linear_model.LinearRegression()
y = df['PM1_OPC_N2_1'].values.reshape(-1,1)
X = df['PM1_GRIMM'].values.reshape(-1,1)
regr.fit(X, y)

a=(regr.coef_[0])
b=(regr.intercept_)
a=float(a)
b=float(b)
df['PM1_OPC_N2_1_pred']=df['PM1_GRIMM']*a + b
y_act = df['PM1_OPC_N2_1'].values.reshape(-1,1)
y_pred = df['PM1_OPC_N2_1_pred'].values.reshape(-1,1)
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_act, y_pred,squared=False))
```

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm
#we can use R style formulas:
results = smf.ols('PM1_OPC_N2_1~PM1_GRIMM', data=df).fit()
print(results.summary())
```

```python
#from sklearn import linear_model
#regr = linear_model.LinearRegression()
#X = df['PM1_OPC_N2_1'].values.reshape(-1,1)
#y = df['PM1_GRIMM'].values.reshape(-1,1)
#regr.fit(X, y)
#a=regr.coef_[0]
#b=regr.intercept_
slope, intercept, r_value, p_value, std_err = stats.linregress(df['PM1_GRIMM'],df['PM1_OPC_N2_1'])
a='{0:0.2f}'.format(intercept)
b='{0:0.2f}'.format(slope)
c='{0:0.2f}'.format(r_value)
g = sns.lmplot('PM1_GRIMM','PM1_OPC_N2_1',df,fit_reg=True,aspect=1.1,ci=None, scatter_kws={"s": 100})
g.set(ylabel='PM$_{1}$ OPC N2 (1) ($\mu g m^{-3}$) ', xlabel='PM$_{1}$ GRIMM ($\mu g m^{-3}$)')
props = dict(boxstyle='round', alpha=0.5,color=sns.color_palette()[0])
#textstr = '$y=1.42 + 1.27x$'
plt.plot([300,800], [300,800], 'r',label='1:1 line')
plt.legend()
textstr='y= {} + {}x , r={}'.format(a,b,c)
g.ax.text(0.4, 1, textstr,transform=g.ax.transAxes, fontsize=14, bbox=props)
```
