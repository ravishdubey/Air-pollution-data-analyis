# Linear regression

# PM1 OPC N2 1 VS GRIMM

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
y = df['PM1_OPC_N2_1'].values.reshape(-1,1)
X = df['PM1_GRIMM'].values.reshape(-1,1)
regr.fit(X, y)
print(regr.coef_[0])
print(regr.intercept_)
```

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm
#we can use R style formulas:
results = smf.ols('PM1_OPC_N2_1~PM1_GRIMM', data=df).fit()
print(results.summary())
```

```python
#from sklearn import linear_model
#regr = linear_model.LinearRegression()
#X = df['PM1_OPC_N2_1'].values.reshape(-1,1)
#y = df['PM1_GRIMM'].values.reshape(-1,1)
#regr.fit(X, y)
#a=regr.coef_[0]
#b=regr.intercept_
slope, intercept, r_value, p_value, std_err = stats.linregress(df['PM1_GRIMM'],df['PM1_OPC_N2_1'])
a='{0:0.2f}'.format(intercept)
b='{0:0.2f}'.format(slope)
c='{0:0.4f}'.format(r_value)

g = sns.lmplot('PM1_GRIMM','PM1_OPC_N2_1',df,fit_reg=True,aspect=1.1,ci=None, scatter_kws={"s": 100})
sns.set(font_scale=1.8, style='whitegrid')
plt.ylabel('PM$_{1}$ OPC N2 (1) ($\mathrm{\mu g \ m^{-3}}$)', fontsize=22, labelpad=20)
plt.xlabel('PM$_{1}$ Grimm ($\mathrm{\mu g \ m^{-3}}$)', fontsize=22, labelpad=20)
props = dict(boxstyle='round', alpha=0.5,color=sns.color_palette()[0])
#textstr = '$y=1.42 + 1.27x$'
c=float(c)
c=c*c
textstr='y= {} + {}x , r$^2$={:0.3f}'.format(a,b,c)
g.ax.text(0, 1.05, textstr,transform=g.ax.transAxes, fontsize=22, bbox=props)
plt.plot([0,800], [0,800], 'r',label='1:1 line')
plt.legend(loc='upper center',bbox_to_anchor=(0.24, 1),ncol=3, fancybox=True, shadow=True)
plt.savefig("fig_tube_regression_pm1_grimm_opcn2_1.png", dpi=300,bbox_inches="tight")
```
